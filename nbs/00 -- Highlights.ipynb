## Imports

```python
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import numpy as np
import pyarrow as pa
import pandas as pd
import seaborn as sns
from matplotlib.ticker import FuncFormatter
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
sns.set()
sns.set_context("poster")

pd.options.display.max_columns = None
pd.options.display.max_rows = None
pd.options.display.precision = 4

dollar_formatter = FuncFormatter(lambda x, pos: f"${x:,.0f}")
thousands_formatter = FuncFormatter(lambda x, pos: f"{x:,.0f}")
string_pa = pd.ArrowDtype(pa.string())
```

## Data

```python
NPOINTS = 500
x = np.linspace(0, 10 * np.pi, num=NPOINTS)
y = np.sin(x)
```

## Functions

```python
def plot_example(title=None, zeroline=False):
    fig, ax = plt.subplots(figsize=(10, 6))
    fig.patch.set_facecolor('w')
    ax.plot(x, y)
    if zeroline:
        ax.axhline(0, c='k')
    if title:
        ax.set_title(f"{title}")
    ax.set_xlabel("Days since the incident")
    # ax.xaxis.set_major_formatter(dollar_formatter)
    fig.tight_layout()
```

## Plots

### How did we do second quarter?

```python
plot_example(title="Here is an example", zeroline=True)
```
