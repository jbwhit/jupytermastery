## script

We all know  `⇧ + ↵` to run a cell and move the cursor to the next cell.

The tip is, if you are creating a plot, it can be really annoying to execute a long cell with code, run it, see the plot and have to edit.

Jarring. So `⌘ + ↵` will run and keep the cell selected, so you don't have to scroll and reselect.

```python
fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('w')
x = np.linspace(0, 2 * np.pi)
y = np.sin(x)





























ax.scatter(x, y)
fig.tight_layout()
```
