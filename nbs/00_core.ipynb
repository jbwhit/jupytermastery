## Keyboard shortcuts

Bring up help: ⇧ + ⌘ + H -- I almost never do this and don't find it helpful. 

Consider bringing up the command palette and typing into there:  ⇧ + ⌘ + C

Which brings me to my first lesson: once you know there's a keyboard shortcut to do something you have to use it. If you look it up, don't press enter on the command palette, force yourself to type the shortcut.

## From now on you never click the run arrow with your mouse

You run ⇧ + ↵ to execute and select the next cell. There's a different shortcut for inserting a new cell below but you don't need that. 

One last one that's useful is ⌘ + ↵ to stay on the same cell -- especially with the ⌃  + R option when doing plots (more on that later).

## Select multiple cells

First Escape (⎋)

## Collapsible output 

Good time to show that there's collapsible output. Click the space, and you will fully expand vs. give a smaller footprint. 

Next, you can click the blue bar itself and get it to go away even more. 

Finally, you can click the blue bar next to the input and really collapse it. 

You can also have sections collapse -- be careful with this one and just use it for collapsing, I've found it's buggy to move collapsed things around. So, don't move collapsed things around.
